```python
Food_Production = Food_Production.copy()
Food_Production['x-mean'] = abs(Food_Production['Value'] - Food_Production['Value'].mean())
Food_Production['1.96*std'] = 1.96*Food_Production['Value'].std()
Food_Production['Outlier'] = abs(Food_Production['Value'] - Food_Production['Value'].mean()) > 1.96*Food_Production['Value'].std()
Food_Production
```

```python
NDF = Food_Production.copy()

year = NDF.groupby('Year')

NDF['Lower'] = year['Value'].transform(lambda x: x.quantile(q=.25) - (1.5*(x.quantile(q=.75) - x.quantile(q=.25))) )
NDF['Upper'] = year['Value'].transform(lambda x: x.quantile(q=.75) + (1.5*(x.quantile(q=.75) - x.quantile(q=.25))) )
NDF['Outlier'] = (NDF['Value'] < NDF['Lower']) | (NDF['Value'] > NDF['Upper'])

NDF
```
